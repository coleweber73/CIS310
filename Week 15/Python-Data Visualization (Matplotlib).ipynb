```python
df = {'col1':[1, 4, 3, 6, 8, 7, 5, 3, 2, 8],
      'col2':[3, 5, 7, 8, 3, 2, 1, 3, 4, 7],
      'col3':[4, 6, 3, 2, 6, 8, 5, 4, 2, 1]}
df = pd.DataFrame(df)
df
```

```python
fig = plt.figure()
ax = fig.add_subplot(1,1,1)
ax.plot(df['col1'], 'k.', label='Column 1')
ax.plot(df['col2'], 'ko--', label='Column 2')
ax.legend(loc='best')
```

```python
# DataFrame Plotting
df['col2'].plot.bar()
```

```python
#Ticks, Labels, and Legends
fig = plt.figure()
ax = fig.add_subplot(1,1,1)
ax.plot(randn(1000).cumsum())
```

```python
ticks = ax.set_xticks([0,250,500,750,1000])
labels = ax.set_xticklabels(['one', 'two','three','four','five'], rotation=30, fontsize='small')
ax.set_title('My Matplotlib Plot')
ax.set_xlabel('Stages')
ax.set_ylabel('Numbers')
fig
```

```python
props = {'title': 'My Matplotlib with Ticks', 
         'xlabel': 'Stages', 
         'ylabel': 'Numbers'}
ax.set(**props)
fig
```

```python
fig = plt.figure()
ax = fig.add_subplot(1,1,1)

ax.plot(randn(1000).cumsum(), 'k', label = "One")
ax.plot(1.3 * randn(1000).cumsum(), 'k--', label = "Two")
ax.plot(5 * randn(1000), 'k.', label = "Three", alpha = 0.3)
ax.legend(loc='best')
```
